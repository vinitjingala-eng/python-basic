4. Write a program that checks if a given word is a Python keyword?

```python
import keyword

def is_python_keyword(word):
  return keyword.iskeyword(word)

word_to_check_1 = "for"
word_to_check_2 = "my_variable"
word_to_check_3 = "True"

if is_python_keyword(word_to_check_1):
  print(f"'{word_to_check_1}' is a Python keyword.")
else:
  print(f"'{word_to_check_1}' is not a Python keyword.")

if is_python_keyword(word_to_check_2):
  print(f"'{word_to_check_2}' is a Python keyword.")
else:
  print(f"'{word_to_check_2}' is not a Python keyword.")

if is_python_keyword(word_to_check_3):
  print(f"'{word_to_check_3}' is a Python keyword.")
else:
  print(f"'{word_to_check_3}' is not a Python keyword.")
```

14. Write a program to reverse a string using a while loop.

```python
def reverse_string_while_loop(input_string):

    reversed_string = ""
    index = len(input_string) - 1

    while index >= 0:
        reversed_string += input_string[index]
        index -= 1

    return reversed_string

original_str = "Hello World!"
reversed_str = reverse_string_while_loop(original_str)
print(f"Original string: {original_str}")
print(f"Reversed string: {reversed_str}")
```

 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop?

```python

try:
    num = int(input("Enter a positive integer: "))
except ValueError:
    print("Invalid input. Please enter an integer.")
else:
    if num < 0:
        print("Factorial is not defined for negative numbers.")
    elif num == 0:
        print("The factorial of 0 is 1.")
    else:
        factorial = 1
        i = 1
        while i <= num:
            factorial *= i
            i += 1
        print(f"The factorial of {num} is {factorial}.")
```
